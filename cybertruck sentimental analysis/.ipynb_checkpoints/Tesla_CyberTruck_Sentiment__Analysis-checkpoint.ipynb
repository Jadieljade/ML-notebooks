## Focus on Key Columns & Relationships

### 'Comment' Column

```python
# Analyzing comment length distribution
df['comment_length'] = df['Comment'].str.len()
df['comment_length'].describe()
```

```python
# Unique words in Comments
unique_words = df['Comment'].str.split().explode().unique()
print(f"Number of unique words: {len(unique_words)}")
```

```python
# Frequent words (excluding stopwords)
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop = stopwords.words('english')
df['Comment_no_stopwords'] = df['Comment'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop]))
frequent_words = df['Comment_no_stopwords'].str.split().explode().value_counts().head(10)
print("Top 10 most frequent words (excluding stopwords):")
print(frequent_words)
```

### 'Published_date' Column

```python
# Ensure 'Published_date' is in datetime format
df['Published_date'] = pd.to_datetime(df['Published_date'])
```

```python
# Range
start_date = df['Published_date'].min().strftime('%Y-%m-%d')
end_date = df['Published_date'].max().strftime('%Y-%m-%d')
print(f"Comments range from: {start_date} to {end_date}")
```

```python
# Distribution over time, aggregated daily
df['Published_date'].dt.date.value_counts().plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Number of Comments')
plt.title('Distribution of Comments Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 'Likes' Column

```python
# Distribution of Likes
plt.hist(df['Likes'])
plt.xlabel('Likes')
plt.ylabel('Number of Comments')
plt.title('Distribution of Likes')
plt.show()
```

### 'Username' Column

```python
# Count Unique Users
n_unique_users = df['Username'].nunique()
print(f"Number of unique users: {n_unique_users}")
```

```python
# Most Prolific Commenters
top_commenters = df['Username'].value_counts().head(10)
print("Top 10 most prolific commenters:")
print(top_commenters)
```

```python
# Calculate what percentage of total comments are made by the top 10% of users.
# 1. Get total number of comments
total_comments = len(df)

# 2. Find top 10% of users based on comment count
top_10_percentile = int(0.1 * df['Username'].nunique())  # How many users make up the top 10%
top_users = df['Username'].value_counts().head(top_10_percentile)

# 3. Calculate the number of comments made by top users
comments_by_top_users = top_users.sum()

# 4. Calculate the percentage
percentage = (comments_by_top_users / total_comments) * 100

print(f"The top 10% of users contribute {percentage:.2f}% of the total comments.")
```

## Data Preparation and Cleaning

```python
# Copy the necessary columns
analysis_df = df[['Published_date', 'Likes', 'Comment']].copy()

# Extract the date component
analysis_df['Published_date'] = analysis_df['Published_date'].dt.date
# Converting 'Published_date' dataype from object to datetime
analysis_df['Published_date'] = pd.to_datetime(analysis_df['Published_date'])

# View the updated dataset
analysis_df.head()
```

## Sentiment Analysis with VADER

```python
analysis_df.dtypes
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
# Calculate sentiment scores
analysis_df['compound'] = analysis_df['Comment'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
# Categorize Sentiment
analysis_df['sentiment'] = analysis_df['compound'].apply(lambda score: 'positive' if score >= 0.05 else ('negative' if score <= -0.05 else 'neutral'))
```

## Analyzing the Sentiment
### Overall Sentiment Distribution:

```python
analysis_df['sentiment'].value_counts().plot(kind='bar')
```

### Sentiment & Likes

```python
analysis_df.groupby('sentiment')['Likes'].mean()
```

### Sentiment over Time

```python
analysis_df.groupby(analysis_df['Published_date'].dt.date)['compound'].mean().plot()
```

## Visualization

### Word Cloud

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS

def generate_wordcloud(text_data, title, mask=None):
    wordcloud = WordCloud(background_color='white', max_words=200, stopwords=STOPWORDS, mask=mask).generate(text_data)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title, fontsize=16)
    plt.show()

# Example usage:
positive_comments = analysis_df[analysis_df['sentiment'] == 'positive']['Comment'].str.cat(sep=' ')
generate_wordcloud(positive_comments, title='Positive Sentiment Word Cloud')
```

### Interactive Visualizations

```python
!pip install plotly
```

```python
import plotly.express as px

df_sentiment_trends = analysis_df.groupby(analysis_df['Published_date'].dt.date)['compound'].mean().reset_index()

fig = px.line(df_sentiment_trends, x='Published_date', y='compound', title='Sentiment Trends over Time')

# Interactive Features:
fig.update_layout(
    updatemenus=[
        dict(buttons=list([
            dict(label="All", method="update", args=[{"visible": [True]}, {"title": "All"}]),
            dict(label="Positive", method="update", args=[{"visible": [True, False, False]}, {"title": "Positive"}]),
            dict(label="Negative", method="update", args=[{"visible": [False, True, False]}, {"title": "Negative"}]),
            dict(label="Neutral", method="update", args=[{"visible": [False, False, True]}, {"title": "Neutral"}])
        ]), direction="down")
    ]
)
fig.show()
```
